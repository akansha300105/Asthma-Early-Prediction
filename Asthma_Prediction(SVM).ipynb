```python
# Importing the dependencies
import pandas as pd
import numpy as np
```

```python
asthma_data = pd.read_csv('asthma_disease_data.csv')
asthma_data.head()
```

```python
asthma_data.shape
```

```python
asthma_data.info()
```

```python
# finding missing values
asthma_data.isnull().sum()
```

```python
# counting the variables based on the label
asthma_data['Diagnosis'].value_counts()
```

0 --> No asthma

1 --> Asthma predicted

```python
asthma_data.drop(columns='DoctorInCharge', axis=1, inplace=True)
```

```python
asthma_data.head()
```

```python
asthma_data.tail()
```

```python
asthma_data_zero=asthma_data[asthma_data['Diagnosis']==0]
asthma_data_one=asthma_data[asthma_data['Diagnosis']==1]
asthma_data_zero=asthma_data_zero.sample(n=124)
asthma_data=pd.concat([asthma_data_zero,asthma_data_one])
asthma_data.shape
```

```python
# separating all the data and the labels
X = asthma_data.drop(columns='Diagnosis', axis=1)
Y = asthma_data['Diagnosis']
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Data Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Training the model

```python
from sklearn import svm
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train_scaled, Y_train)
```

Model evaluation and accuracy score

```python
# accuracy score on training data
from sklearn.metrics import accuracy_score
X_train_prediction = classifier.predict(X_train_scaled)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
# Accuracy score on test data
Y_train_prediction = classifier.predict(X_test_scaled)
test_data_accuracy = accuracy_score(Y_test, Y_train_prediction)
```

```python
print('Accuracy score of the testing data :', test_data_accuracy)
```

```python
# classification report
from sklearn.metrics import classification_report
y_pred=classifier.predict(X_test_scaled)
print(classification_report(Y_test,y_pred))
```
