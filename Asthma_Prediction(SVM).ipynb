```python
# Importing the dependencies
import pandas as pd
import numpy as np
```

```python
asthma_data = pd.read_csv('asthma_disease_data.csv')
asthma_data.head()
```

```python
asthma_data.shape
```

```python
asthma_data.info()
```

```python
# finding missing values
asthma_data.isnull().sum()
```

```python
# counting the variables based on the label
asthma_data['Diagnosis'].value_counts()
```

0 --> No asthma

1 --> Asthma predicted

```python
# separating the data for analysis
no_as = asthma_data[asthma_data.Diagnosis == 0]
yes_as = asthma_data[asthma_data.Diagnosis == 1]
```

```python
print(no_as.shape)
print(yes_as.shape)
```

```python
no_as_sample = no_as.sample(n=124)
```

```python
no_as_sample.shape
```

```python
# concatenate the two dataframe
asth_data_new = pd.concat([no_as_sample, yes_as], axis=0)
```

```python
asth_data_new.head()
```

```python
asth_data_new.tail()
```

```python
asth = asth_data_new.drop(columns='DoctorInCharge', axis=1)
```

```python
asth.shape
```

```python
# separating all the data and the labels
X = asth.drop(columns='Diagnosis', axis=1)
Y = asth['Diagnosis']
```

Data Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
print(standardized_data)
```

```python
X = standardized_data
Y = asth['Diagnosis']
```

```python
print(X)
```

```python
print(Y)
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Training the model

```python
from sklearn import svm
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

Model evaluation and accuracy score

```python
# accuracy score on training data
from sklearn.metrics import accuracy_score
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
# Accuracy score on test data
Y_train_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, Y_train_prediction)
```

```python
print('Accuracy score of the testing data :', test_data_accuracy)
```

Making a predictive system

```python
input_data = (7411,5,1,1,2,35.47908018703524,0,4.571439525578311,4.228623350807997,7.42854156483522,1.6666505433422363,8.374471180827463,2.225013650691788,0,0,1,0,0,0,2.6759309813857843,1.7377074458651913,0,1,0,1,0,1)
# changing the input data into array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardize the input data
std_data = scaler.transform(input_data_reshaped)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
    print('The person does not have asthma disease')
else:
    print('The person has asthma disease')
```
